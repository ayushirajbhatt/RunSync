```python
from spotify_recommender import get_recs
from time import sleep
import pandas as pd
import json
```

```python
df = pd.read_csv('Bio_Sensory_Zone_Model_test_1.csv')
df.head(5)
```

```python
def get_song_filters(row):
    filters = {
        'target_tempo': row.recommended_tempo,
        'target_valence': row.recommended_valence,
        'target_popularity': row.recommended_popularity,
        'seed_genres': [json.loads(row.recommended_genre)['name']]
    }
    return filters

df['song_filters'] = df.apply(get_song_filters, axis=1)
```

```python
## Create Playlist 

current_songs = set()

def get_song(row):
    song = get_recs(row.song_filters, current_songs)
    current_songs.add(song['uri'])
    return song

df['song'] = df.apply(get_song, axis=1)
df.head(5)
```

```python
df.to_csv('playlist.csv')
```
